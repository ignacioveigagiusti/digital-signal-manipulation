```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import scipy
import scipy.io.wavfile as wav
import fftplot as fftplt
from scipy.fftpack import fft, ifft
from matplotlib.pyplot import figure
import generador as gen
import IPython.display as ipd
```


Las primeras variables a definir son los límites temporales y la frecuencia de muestreo para nuestro vector temporal. Es fundamental ejecutar el siguiente bloque antes que cualquier otro.


```python
samplerate = 8000  #Frecuencia de Muestreo
timestart=0       #Tiempo Inicial
timestop=1         #Tiempo Final
timestep=1/samplerate           
timevec = np.arange(timestart,timestop,timestep)   #Vector Temporal
N=1024          #Cantidad de puntos de la FFT
```

## Sección 4: Filtraje

En esta sección se pueden aplicar distintos filtros a señales digitales y ver sus efectos.

### Lectura y muestreo de archivos .wav

```python
# importamos las señales Voz masculina y Voz femenina
samplerateVM, sigVM = wav.read('Voz masculina.wav')
samplerateVF, sigVF = wav.read('Voz femenina.wav')

sigVM=signal.decimate(sigVM, 6, n=None, ftype='iir', axis=- 1, zero_phase=True)
sigVF=signal.decimate(sigVF, 6, n=None, ftype='iir', axis=- 1, zero_phase=True)

duracionVM = np.size(sigVM) / samplerate  # operación para obtener la duración de la señal
timevecVM = np.linspace(0, duracionVM, np.size(sigVM))  # vector tiempo
duracionVF = np.size(sigVF) / samplerate  # operación para obtener la duración de la señal
timevecVF = np.linspace(0, duracionVF, np.size(sigVF))  # vector tiempo

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(15,5))
ax1.plot(timevecVM,sigVM)
ax2.plot(timevecVF,sigVF)
ax1.set_title('Voz Masculina')
ax2.set_title('Voz Femenina')
```

### Chirp y Suma de Senoidales

```python
flimchirp=400
chirp = signal.chirp(t=timevec, f0=0, t1=1, f1=flimchirp, method='linear', phi=0, vertex_zero=True)

figure(num=None, figsize=(15, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(timevec,chirp)
plt.title('Chirp')
```

```python
f1=samplerate/200

sin1=gen.generador(timevec, samplerate, 1, 0, f1, 0, 50, 0.5, 1, 0.5, 1)
sin2=gen.generador(timevec, samplerate, 1, 0, 2*f1, 0, 50, 0.5, 1, 0.5, 1)
sin3=gen.generador(timevec, samplerate, 1, 0, 3*f1, 0, 50, 0.5, 1, 0.5, 1)
sin4=gen.generador(timevec, samplerate, 1, 0, 4*f1, 0, 50, 0.5, 1, 0.5, 1)
sin5=gen.generador(timevec, samplerate, 1, 0, 5*f1, 0, 50, 0.5, 1, 0.5, 1)

sinsum=sin1+sin2+sin3+sin4+sin5

figure(num=None, figsize=(15, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(timevec,sinsum)
plt.title('Suma de Senoidales')
```

# Filtros

```python
fc = 200  # frecuencia de corte del filtro
wc = fc / (0.5*samplerate)  # frecuencia normalizada para el filtro digital

# se obtienen los coeficientes de la función de transferencia del filtro
# Butterworth lowpass
b1, a1 = signal.butter(N=3, Wn=wc, btype='low')

# Chebyshev 1 Lowpass
b2, a2 = signal.cheby1(N=3, rp=4, Wn=wc, btype='low')

# Chebyshev 2 Lowpass
b3, a3 = signal.cheby2(N=3, rs=4, Wn=wc, btype='low')

# Elíptico Lowpass
b4, a4 = signal.ellip(N=3, rp=4, rs=40, Wn=wc, btype='low')

# Butterworth Highpass
b5, a5 = signal.butter(N=3, Wn=wc, btype='highpass')

# Chebyshev 1 Highpass
b6, a6 = signal.cheby1(N=3, rp=4, Wn=wc, btype='highpass')

# Chebyshev 2 Highpass
b7, a7 = signal.cheby2(N=3, rs=4, Wn=wc, btype='highpass')

# Elíptico Highpass
b8, a8 = signal.ellip(N=3, rp=4, rs=40, Wn=wc, btype='highpass')

# Butterworth Bandpass
b9, a9 = signal.butter(N=3, Wn=[150/samplerate, 2000/samplerate], btype='band')

# Chebyshev 1 Bandpass
b10, a10 = signal.cheby1(N=3, rp=4, Wn=[150/samplerate, 2000/samplerate], btype='band')

# Chebyshev 2 Bandpass
b11, a11 = signal.cheby2(N=3, rs=4, Wn=[150/samplerate, 2000/samplerate], btype='band')

# Elíptico Bandpass
b12, a12 = signal.ellip(N=3, rp=4, rs=40, Wn=[150/samplerate, 2000/samplerate], btype='band')

# Butterworth Bandstop
b13, a13 = signal.butter(N=3, Wn=[150/samplerate, 2000/samplerate], btype='bandstop')

# Chebyshev 1 Bandstop
b14, a14 = signal.cheby1(N=3, rp=4, Wn=[150/samplerate, 2000/samplerate], btype='bandstop')

# Chebyshev 2 Bandstop
b15, a15 = signal.cheby2(N=3, rs=4, Wn=[150/samplerate, 2000/samplerate], btype='bandstop')

# Elíptico Bandstop
b16, a16 = signal.ellip(N=3, rp=4, rs=40, Wn=[150/samplerate, 2000/samplerate], btype='bandstop')

```

### Voz Masculina - Filtro Butterworth Lowpass

```python
sig_filt = signal.filtfilt(b1, a1, sigVM)
sig_filt_wav_1 = sig_filt.astype(np.int16)
wav.write('VM-Butterlow', samplerateVM, sig_filt_wav_1)

Xo= fft(sigVM,N)
X = fft(sig_filt,N)
lim = int(np.ceil((N+1)/2) - 1)
Xo = np.append(X[lim:1:-1], X[0:lim])
X = np.append(X[lim:1:-1], X[0:lim])
MXo = np.abs(Xo)
MXo = MXo / np.size(MX)
MX = np.abs(X)
MX = MX / np.size(MX)
f =np.linspace((-N/2), (N/2), N-1) * samplerate/N

w1, h1 = signal.freqz(b1,a1)


respuesta_fase_butt = np.unwrap(np.angle(h1))

t,y=signal.impulse2((a1,b1))

#fig, ((ax1, ax2), (ax3, ax4),(ax5,ax6)) = plt.subplots(3,2,figsize=(20,17))
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2,figsize=(20,17))
plt.setp(ax1, xlabel='Tiempo')
plt.setp(ax2, xlabel='Frecuencia')
#plt.setp(ax3, xlabel='Frecuencia')
plt.setp(ax1, ylabel='Amplitud')
plt.setp(ax2, ylabel='Amplitud')
#plt.setp(ax3, ylabel='Amplitud')
#ax1.set_title('Señal Original')
ax2.set_title('Señal Original y Filtrada en Frecuencia')
ax1.set_title('Señal Original y Filtrada en Tiempo')
ax3.set_title('Magnitud de H')
ax4.set_title('Fase de H')
#ax3.set_title('Señal Filtrada en Frecuencia')
#ax4.set_title('Magnitud de H')
#ax5.set_title('Fase de H')
#ax6.set_title('Respuesta al impulso')
#ax1.plot(timevecVM,sigVM)
#ax2.plot(timevecVM,sig_filt)
ax1.plot(timevecVM,sigVM)
ax1.plot(timevecVM,sig_filt)
#ax2.plot(f,MXo)  #-------------------------GRAFICAR SEÑAL SIGVM EN FRECUENCIA SIN FILTRAR, CREO QUE ESTÁ MAL
ax2.plot(f,MX)
ax3.semilogx(w1, 20 * np.log10(abs(h1)), label='butter-lowpass')
ax4.plot(w1, respuesta_fase_butt, label = 'butter')
figure()
plt.plot(t,y, label='Respuesta al impultso h')
plt.legend()
#ax4.semilogx(w1, 20 * np.log10(abs(h1)), label='butter-lowpass')
#ax5.plot(w1, respuesta_fase_butt, label = 'butter')
#ax6.plot(t,y)
```

```python
ipd.Audio(sigVM, rate=samplerate)
```

```python
ipd.Audio(sig_filt_wav_1, rate=samplerate)
```

```python
sig_filt = signal.filtfilt(b2, a2, sigVF)
sig_filt_wav_2 = sig_filt.astype(np.int16)
wav.write('VF-Cheb1low', samplerateVF, sig_filt_wav_2)

X = fft(sig_filt,N)
lim = int(np.ceil((N+1)/2) - 1)
X = np.append(X[lim:1:-1], X[0:lim])
MX = np.abs(X)
MX = MX / np.size(MX)
f =np.linspace((-N/2), (N/2), N-1) * samplerate/N

w2, h2 = signal.freqz(b2,a2)


respuesta_fase_cheby1 = np.unwrap(np.angle(h2))

t,y=signal.impulse2((a2,b2))

fig, ((ax1, ax2), (ax3, ax4),(ax5,ax6)) = plt.subplots(3,2,figsize=(20,17))
plt.setp(ax1, xlabel='Tiempo')
plt.setp(ax2, xlabel='Tiempo')
plt.setp(ax3, xlabel='Frecuencia')
plt.setp(ax1, ylabel='Amplitud')
plt.setp(ax3, ylabel='Amplitud')
ax1.set_title('Señal Original')
ax2.set_title('Señal Filtrada')
ax3.set_title('Señal Filtrada en Frecuencia')
ax4.set_title('Magnitud de H')
ax5.set_title('Fase de H')
ax6.set_title('Respuesta al impulso')
ax1.plot(timevecVF,sigVF)
ax2.plot(timevecVF,sig_filt)
ax3.plot(f,MX)
ax4.semilogx(w2, 20 * np.log10(abs(h2)), label='cheby1-lowpass')
ax5.plot(w2, respuesta_fase_cheby1, label = 'cheby1')
ax6.plot(t,y)
```
